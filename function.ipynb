# **Условие** #
## *Данна функция:* ##
*f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30*

*  *Определить корни*
*  *Найти интервалы, на которых функция возрастает*
*  *Найти интервалы, на которых функция убывает*
*  *Построить график*
*  *Вычислить вершину*
*  *Определить промежутки, на котором f > 0*
*  *Определить промежутки, на котором f < 0*

```python
from sympy import *
from sympy.plotting import plot
from matplotlib import *
init_printing()
x = Symbol('x')
f = -12*x*x*x*x*sin(cos(x)) - 18*x*x*x + 5*x*x + 10*x - 30
plot(f, (x, -100, 100), title = 'График функции')
```

## Поиск асимптот, корней, экстремумов ##

```python
print(f'Горизонтальная асимптота слева: {limit(f, x, -oo)}')
print(f'Горозонтальная асимптота справа: {limit(f, x, oo)}')
print(f'Точка пересечения с осью Y; [0, {f.subs(x, 0)}]')
print(f'Функция f(x) = {f}, имеет бесконечное число корней и не имеет экстремумов')
```

## Наохождение производной ##

```python
print(f'Производная первого порядка: {f.diff(x)}')
```

## График производной функции, интервалы возрастания и убывания функции ##

```python
df = f.diff(x)
print('На участках где: \n(df < 0) - функция убывает \n(df > 0) - функция возрастает')
plot(f, (x, -25, 25), line_color = 'red', title = 'График f(x)')
plot(df, (x, -25, 25), line_color = 'green', title = 'График df(x)')
```
